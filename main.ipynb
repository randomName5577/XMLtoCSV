```python
import xml.etree.ElementTree as ET
```

```python
tree = ET.parse("rows.xml")
```

```python
root = tree.getroot()
```

```python
root.tag
```

```python
root.attrib
```

```python
elemtree = {}
```

```python
i = 0 
for child in root[0].findall('row'):
    element={"Childtag": child.tag}
    #print(child.tag)
    for k,v in child.attrib.items():
        element[k]=v
        #print(k," : ",v)
    for elem in child:
        element[elem.tag]=child.find(elem.tag).text
        #print(elem.tag," : ", child.find(elem.tag).text)
    #print ('\n')
    i +=1
    elemtree[i]=element
```

```python
import csv
```

```python
with open('test.csv', 'w') as f:
    for k in elemtree.keys():
        f.write(f"{k},{elemtree[k]} \n")
    
```
